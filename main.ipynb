```python
import pandas as pd
from generator import generar_reporte

df_encuesta = pd.read_excel("data/bdd-bruta-2.xlsx")
df_encuesta["EMAIL"] = df_encuesta["EMAIL"].str.split("@").str[0]
df_encuesta_2022 = pd.read_excel("data/respuestas-brutas-2022.xlsx")
df_universo = pd.read_excel("data/universo-radiografia-final-2.xlsx")
df_universo["JORNADA"] = df_universo["JORNADA"].astype(str).str.strip().str.capitalize()
df_estres_2022 = pd.read_excel("data/bdd_estres_2022.xlsx")
df_salud_mental_2022 = pd.read_excel("data/2022_subeje_salud_mental.xlsx")
df_bienestar_2022 = pd.read_excel("data/2022_eje_bienestar.xlsx")

df_global = pd.merge(df_encuesta, df_universo, on="EMAIL", how="inner")

### DATOS PARTICIAPANTES POR ESCUELAS Y SEDES/CAMPUS
columnas_participantes = ["SEDE", "ESCUELA", "GENERO", "JORNADA", "TIPO_ALUMNO"]
df_participantes = df_global[columnas_participantes]

### DATOS ESTRES PERCIBIDO
columnas_estres = ["PSS4[PSS01]", "PSS4[PSS02]", "PSS4[PSS03]", "PSS4[PSS04]", "EMAIL"]
df_estres = df_encuesta[columnas_estres].copy()
df_estres["EMAIL"] = df_estres["EMAIL"].str.split("@").str[0]
df_estres = pd.merge(df_universo, df_estres, on="EMAIL", how="inner")

### DATOS ESTRES PERCIBIDO 2022
df_estres_2022 = df_estres_2022.rename(columns={df_estres_2022.columns[0]: "Grupo"})
for col in df_estres_2022.columns[1:]:
    df_estres_2022[col] = pd.to_numeric(df_estres_2022[col], errors="coerce")
df_estres_2022 = df_estres_2022.set_index("Grupo").round(2).T.to_dict()
df_estres_2022 = pd.DataFrame(df_estres_2022)

### DATOS ESTRESORES
columnas_estresores = [
  "ESTRE[ESTRE01]", "ESTRE[ESTRE02]", "ESTRE[ESTRE03]", "ESTRE[ESTRE04]", "ESTRE[ESTRE05]",
  "ESTRE[ESTRE06]", "ESTRE[ESTRE07]", "ESTRE[ESTRE08]", "ESTRE[ESTRE09]", "ESTRE[ESTRE10]",
  "ESTRE[ESTRE11]", "ESTRE[ESTRE12]", "ESTRE[ESTRE13]", "ESTRE[ESTRE14]", "ESTRE[ESTRE15]",
  "ESTRE[ESTRE16]", "ESTRE[ESTRE17]", "ESTRE[ESTRE18]", "ESTRE[ESTRE19]", "ESTRE[ESTRE20]",
  "ESTRE[ESTRE21]", "ESTRE[ESTRE22]", "EMAIL"
]
df_estresores = df_encuesta[columnas_estresores].copy()
df_estresores["EMAIL"] = df_estresores["EMAIL"].str.split("@").str[0]
df_estresores = pd.merge(df_universo, df_estresores, on="EMAIL", how="inner")

### DATOS EXIGENCIA CARGA ACADEMICA
columnas_estresor_exig = ["ESTRE[ESTRE18]", "EXIG[1]", "EXIG[2]", "EXIG[3]", "EXIG[4]", "EXIG[5]", "EXIG[6]", "EMAIL"]
df_exigencia = df_encuesta[columnas_estresor_exig].copy()
df_exigencia["EMAIL"] = df_exigencia["EMAIL"].str.split("@").str[0]
df_exigencia = pd.merge(df_universo, df_exigencia, on="EMAIL", how="inner")

### DATOS ESTRATEGIAS DE AFRONTAMIENTO
columnas_afrontamiento = ["ESTRA[ESTRA01]", "ESTRA[ESTRA02]", "ESTRA[ESTRA03]", "ESTRA[ESTRA04]", "ESTRA[ESTRA05]", "ESTRA[ESTRA06]", "ESTRA[ESTRA07]", "EMAIL"]
df_afrontamiento = df_encuesta[columnas_afrontamiento].copy()
df_afrontamiento["EMAIL"] = df_afrontamiento["EMAIL"].str.split("@").str[0]
df_afrontamiento = pd.merge(df_universo, df_afrontamiento, on="EMAIL", how="inner")

### DATOS PROCRASTINACION
columnas_procrastinacion = ["PROCAS[PROCA]", "EMAIL"]
df_procrastinacion = df_encuesta[columnas_procrastinacion].copy()
df_procrastinacion["EMAIL"] = df_procrastinacion["EMAIL"].str.split("@").str[0]
df_procrastinacion = pd.merge(df_universo, df_procrastinacion, on="EMAIL", how="inner")

### DATOS CONDICION DE SALUD CRONICA
columnas_salud_cronica = ["SALFIS01", "SALFIS02", "EMAIL"]
df_salud_cronica = df_encuesta[columnas_salud_cronica].copy()
df_salud_cronica["EMAIL"] = df_salud_cronica["EMAIL"].str.split("@").str[0]
df_salud_cronica = pd.merge(df_universo, df_salud_cronica, on="EMAIL", how="inner")

### PROFUNDIZACION DEPRESION
columnas_depresion = ["PHQ8[PHQ01]", "PHQ8[PHQ02]", "PHQ8[PHQ03]", "PHQ8[PHQ04]", "PHQ8[PHQ05]", "PHQ8[PHQ06]", "PHQ8[PHQ07]", "PHQ8[PHQ08]", "EMAIL", "ED15[ED01]", "ED15[ED02]"] + ["DSM5[DSM" + str(i).zfill(2) + "]" for i in range(1, 23)]
df_depresion = df_encuesta[columnas_depresion].copy()
df_depresion["EMAIL"] = df_depresion["EMAIL"].str.split("@").str[0]
df_depresion = pd.merge(df_universo, df_depresion, on="EMAIL", how="inner")

### PROFUNDIZACION ANSIEDAD
columnas_ansiedad = ["GAD7[GAD01]", "GAD7[GAD02]", "GAD7[GAD03]", "GAD7[GAD04]", "GAD7[GAD05]", "GAD7[GAD06]", "GAD7[GAD07]", "EMAIL", "ED15[ED01]", "ED15[ED02]"] + ["DSM5[DSM" + str(i).zfill(2) + "]" for i in range(1, 23)]
df_ansiedad = df_encuesta[columnas_ansiedad].copy()
df_ansiedad["EMAIL"] = df_ansiedad["EMAIL"].str.split("@").str[0]
df_ansiedad = pd.merge(df_universo, df_ansiedad, on="EMAIL", how="inner")

### PROFUNDIZACION PERSONALIDAD
columnas_personalidad = ["LPFS[LPFS01]", "LPFS[LPFS02]", "LPFS[LPFS03]", "LPFS[LPFS04]", "LPFS[LPFS05]", "LPFS[LPFS06]", "LPFS[LPFS07]", "LPFS[LPFS08]", "LPFS[LPFS09]","LPFS[LPFS10]", "LPFS[LPFS11]", "LPFS[LPFS12]", "EMAIL", "ED15[ED01]", "ED15[ED02]"] + ["DSM5[DSM" + str(i).zfill(2) + "]" for i in range(1, 23)]
df_personalidad = df_encuesta[columnas_personalidad].copy()
df_personalidad["EMAIL"] = df_personalidad["EMAIL"].str.split("@").str[0]
df_personalidad = pd.merge(df_universo, df_personalidad, on="EMAIL", how="inner")

### PROFUNDIZACION MARIHUANA
columnas_marihuana = ["CAST[CAST01]", "CAST[CAST02]", "CAST[CAST03]", "CAST[CAST04]", "CAST[CAST05]", "CAST[CAST06]", "EMAIL", "ED15[ED01]", "ED15[ED02]"] + ["DSM5[DSM" + str(i).zfill(2) + "]" for i in range(1, 23)]
df_marihuana = df_encuesta[columnas_marihuana].copy()
df_marihuana["EMAIL"] = df_marihuana["EMAIL"].str.split("@").str[0]
df_marihuana = pd.merge(df_universo, df_marihuana, on="EMAIL", how="inner")

### PROFUNDIZACOIN ALCOHOL
columnas_alcohol = ["AUDIT01[AUDIT01]", "AUDIT02[AUDIT02]", "AUDIT38[AUDIT03]", "AUDIT38[AUDIT04]", "AUDIT38[AUDIT05]", "AUDIT38[AUDIT06]", "AUDIT38[AUDIT07]", "AUDIT38[AUDIT08]", "AUDIT910[AUDIT09]", "AUDIT910[AUDIT10]", "EMAIL", "ED15[ED01]", "ED15[ED02]"] + ["DSM5[DSM" + str(i).zfill(2) + "]" for i in range(1, 23)]
df_alcohol = df_encuesta[columnas_alcohol].copy()
df_alcohol["EMAIL"] = df_alcohol["EMAIL"].str.split("@").str[0]
df_alcohol = pd.merge(df_universo, df_alcohol, on="EMAIL", how="inner")

### PROFUNDIZACION SUICIDIO
columnas_suicidio = ["PHQ09[PHQ09F]", "EMAIL", "ED15[ED01]", "ED15[ED02]", "CSS01", "CSS02", "CSS03", "CSS04", "CSS05", "CSS06"] + ["DSM5[DSM" + str(i).zfill(2) + "]" for i in range(1, 23)]
df_suicidio = df_encuesta[columnas_suicidio].copy()
df_suicidio["EMAIL"] = df_suicidio["EMAIL"].str.split("@").str[0]
df_suicidio = pd.merge(df_universo, df_suicidio, on="EMAIL", how="inner")

### SALUD MENTAL GRUPO VERDE
columnas_grupo_verde = ["DSM5[DSM" + str(i).zfill(2) + "]" for i in range(1, 23)] + ["ED15[ED01]", "ED15[ED02]", "EMAIL", "PHQ09[PHQ09F]"]
df_grupo_verde = df_encuesta[columnas_grupo_verde].copy()
df_grupo_verde["EMAIL"] = df_grupo_verde["EMAIL"].str.split("@").str[0]
df_grupo_verde = pd.merge(df_universo, df_grupo_verde, on="EMAIL", how="inner")

# Keep only rows where all DSM5 items and ED15[ED01], ED15[ED02] have a non-empty response
required_cols = ["DSM5[DSM" + str(i).zfill(2) + "]" for i in range(1, 23)] + ["ED15[ED01]", "ED15[ED02]"]
mask = df_grupo_verde[required_cols].notna() & (df_grupo_verde[required_cols].astype(str).apply(lambda s: s.str.strip()) != "")
df_grupo_verde = df_grupo_verde[mask.all(axis=1)].reset_index(drop=True)

### SALUD MENTAL GRUPO ROJO
columnas_grupo_rojo = list(set(columnas_depresion + columnas_ansiedad + columnas_personalidad + columnas_marihuana + columnas_alcohol + columnas_suicidio))
df_grupo_rojo = df_encuesta[columnas_grupo_rojo].copy()
df_grupo_rojo["EMAIL"] = df_grupo_rojo["EMAIL"].str.split("@").str[0]
df_grupo_rojo = pd.merge(df_universo, df_grupo_rojo, on="EMAIL", how="inner")

### PERCEPCI√ìN DE NECESIDAD Y ACCESO A TRATAMIENTO
columnas_necesidad_tratamiento = ["ACCTTO", "NECTTO", "LUGTTO", "TIPTTO", "EMAIL"]
df_necesidad_tratamiento = df_encuesta[columnas_necesidad_tratamiento].copy()
df_necesidad_tratamiento["EMAIL"] = df_necesidad_tratamiento["EMAIL"].str.split("@").str[0]
df_necesidad_tratamiento = pd.merge(df_universo, df_necesidad_tratamiento, on="EMAIL", how="inner")

### BARRERAS ACCESO A TRATAMIENTO
columnas_barreras = ["BARR[BARR01]", "BARR[BARR02]", "BARR[BARR03]", "BARR[BARR04]", "BARR[BARR05]", "BARR[BARR06]", "BARR[BARR07]", "BARR[BARR08]", "BARR[BARR09]", "EMAIL"]
df_barreras = df_encuesta[columnas_barreras].copy()
df_barreras["EMAIL"] = df_barreras["EMAIL"].str.split("@").str[0]
df_barreras = pd.merge(df_universo, df_barreras, on="EMAIL", how="inner")

### APOYO PERCIBIDO
columnas_apoyo = ["APOYO[APOY01]", "APOYO[APOY02]", "APOYO[APOY03]", "APOYO[APOY04]", "APOYO[APOY05]", "APOYO[APOY06]", "EMAIL"]
df_apoyo = df_encuesta[columnas_apoyo].copy()
df_apoyo["EMAIL"] = df_apoyo["EMAIL"].str.split("@").str[0]
df_apoyo = pd.merge(df_universo, df_apoyo, on="EMAIL", how="inner")

### FLOW STATES
columnas_flow_states = ["FLOW[FLOW01]", "FLOW[FLOW02]", "FLOW[FLOW03]", "FLOW[FLOW04]", "FLOW[FLOW05]", "FLOW[FLOW06]", "FLOW[FLOW07]", "FLOW[FLOW08]", "FLOW[FLOW09]", "FLOW[FLOW10]", "FLOW[FLOW11]", "FLOW[FLOW12]", "EMAIL"]
df_flow_states = df_encuesta[columnas_flow_states].copy()
df_flow_states["EMAIL"] = df_flow_states["EMAIL"].str.split("@").str[0]
df_flow_states = pd.merge(df_universo, df_flow_states, on="EMAIL", how="inner")

### ACTIVIDAD FISICA
columnas_actividad_fisica = ["ACTFIS01[ACTFIS01]", "EMAIL"]
df_act_fisica = df_encuesta[columnas_actividad_fisica].copy()
df_act_fisica["EMAIL"] = df_act_fisica["EMAIL"].str.split("@").str[0]
df_act_fisica = pd.merge(df_universo, df_act_fisica, on="EMAIL", how="inner")

### ALIMENTACION
columnas_alimentacion = ["ALIMEN[ALIMEN01]", "ALIMEN[ALIMEN02]", "ALIMEN[ALIMEN03]", "ALIMEN[ALIMEN04]", "EMAIL"]
df_alimentacion = df_encuesta[columnas_alimentacion].copy()
df_alimentacion["EMAIL"] = df_alimentacion["EMAIL"].str.split("@").str[0]
df_alimentacion = pd.merge(df_universo, df_alimentacion, on="EMAIL", how="inner")

### INGESTA LIQUIDOS
columnas_ing_liquidos = ["INGLIQ01", "EMAIL"]
df_ing_liquidos = df_encuesta[columnas_ing_liquidos].copy()
df_ing_liquidos["EMAIL"] = df_ing_liquidos["EMAIL"].str.split("@").str[0]
df_ing_liquidos = pd.merge(df_universo, df_ing_liquidos, on="EMAIL", how="inner")

### TABAQUISMO
columnas_tabaco = ["DSM5[DSM21]", "EMAIL"]
df_tabaco = df_encuesta[columnas_tabaco].copy()
df_tabaco["EMAIL"] = df_tabaco["EMAIL"].str.split("@").str[0]
df_tabaco = pd.merge(df_universo, df_tabaco, on="EMAIL", how="inner")

### NIVEL BIENESTAR 
perma_cols = [f"PERMA1[PER{i:02d}]" for i in range(1, 21)] + ["PERMA21", "PERMA22", "PERMA23", "EMAIL"]
df_perma = df_encuesta[perma_cols].copy()
df_perma["EMAIL"] = df_perma["EMAIL"].str.split("@").str[0]
df_perma = pd.merge(df_universo, df_perma, on="EMAIL", how="inner")

generar_reporte(df_global, df_participantes, df_estres, df_estres_2022, df_estresores, 
                df_exigencia, df_afrontamiento, df_procrastinacion, df_salud_cronica, df_depresion, df_ansiedad, df_personalidad, df_marihuana, 
                df_alcohol, df_suicidio, df_grupo_verde, df_grupo_rojo, df_necesidad_tratamiento, df_barreras, df_apoyo, df_flow_states, df_act_fisica,
                df_alimentacion, df_ing_liquidos, df_tabaco, df_perma, df_salud_mental_2022, df_bienestar_2022)
```
